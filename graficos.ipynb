```python
import numpy as np

anios = []
for i in range(2011, 2021, 1):
    anios.append(i)

nivel1 = np.random.rand(10)*100
nivel2 = np.random.rand(10)*200 + 100
nivel3 = np.random.rand(10)*300 + 200
nivel4 = np.random.rand(10)*400 + 300
nivel5 = np.random.rand(10)*500 + 400

plt.plot(anios, nivel1, label="Nivel 1", color="purple", marker="<", linestyle="-")
plt.plot(anios, nivel2, label="Nivel 2", color="red", marker="*", linestyle="--")
plt.plot(anios, nivel3, label="Nivel 3", color="blue", marker="^", linestyle="-.")
plt.plot(anios, nivel4, label="Nivel 4", color="black", marker=".", linestyle=":")
plt.plot(anios, nivel5, label="Nivel 5", color="green", marker="+", linestyle=" ")
plt.legend()
plt.title("Examen de Certificacion")
plt.xlabel("Anios que se ha realizado el examen")
plt.ylabel("Puntaje")
plt.yticks(np.arange(0, 1001, 50))
plt.grid()
plt.minorticks_on()

plt.show()

```

```python
eje_x = np.arange(0, 10)

plt.bar(eje_x, nivel5, width=1/5)
plt.bar(eje_x+0.2, nivel4, width=1/5)
plt.bar(eje_x+0.4, nivel3, width=1/5)
plt.bar(eje_x+0.6, nivel2, width=1/5)
plt.bar(eje_x+0.8, nivel1, width=1/5)

plt.show()
```

```python

plt.barh(eje_x+0.8, nivel5, height=1/5)
plt.barh(eje_x+0.6, nivel4, height=1/5)
plt.barh(eje_x+0.4, nivel3, height=1/5)
plt.barh(eje_x+0.2, nivel2, height=1/5)
plt.barh(eje_x, nivel1, height=1/5)
```

```python
plt.bar(eje_x, nivel5, width=1/5)
plt.bar(eje_x, nivel4, bottom=nivel5)
plt.bar(eje_x, nivel3, bottom=nivel4+nivel5)
plt.bar(eje_x, nivel2, bottom=nivel3+nivel4+nivel5)
plt.bar(eje_x, nivel1, bottom=nivel2+nivel3+nivel4+nivel5)

plt.show()
```

```python
plt.scatter(eje_x, nivel1)
plt.scatter(eje_x, nivel2)
plt.scatter(eje_x, nivel3)
plt.scatter(eje_x, nivel4)
plt.scatter(eje_x, nivel5)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

pinguinos = sns.load_dataset("penguins")
pinguinos.columns = ["especie", "isla", "longitud_pico", "profundidad_pico",
                     "longitud_aleta", "masa", "sexo"]
pinguinos
```

```python
sns.relplot(data=pinguinos)
plt.show()
```

```python
sns.set_theme(style="darkgrid", context="talk")
relacional = sns.relplot(data=pinguinos,
                         height=8,
                         aspect=1.3,
                         markers=["s", ".", "P", "d"])
relacional.set(title="Analisis de Pinguinos")
relacional.set_ylabels("Variables")
relacional.set_xlabels("Numero de Pinguino")

plt.show()
```

```python
sns.relplot(data=pinguinos,
            x="longitud_pico",
            y="longitud_aleta")

sns.relplot(data=pinguinos,
            x="longitud_pico",
            y="longitud_aleta",
            hue="masa",
            s=300)


sns.relplot(data=pinguinos,
            x="longitud_pico",
            y="longitud_aleta",
            hue="especie",
            s=300)


sns.relplot(data=pinguinos,
            x="longitud_pico",
            y="longitud_aleta",
            hue="especie",
            row="sexo",
            size="masa",
            sizes= (100, 500),
            style="sexo")

sns.relplot(data=pinguinos,
            x="longitud_pico",
            y="longitud_aleta",
            hue="especie",
            col="isla",
            row="sexo",
            size="masa",
            sizes= (100, 500),
            style="sexo")

plt.show()
```

```python
sns.relplot(data=pinguinos,
            x="especie",
            y="masa",
            kind="line",
            hue="sexo",
            style="sexo")

sns.relplot(data=pinguinos,
            x="profundidad_pico",
            y="longitud_aleta",
            kind="line",
            hue="especie",
            style="especie")

plt.show()
```

```python
figura, ejes = plt.subplots(1, 2, figsize=(15, 5))

sns.lineplot(data=pinguinos,
            x="especie",
            y="masa",
            hue="sexo",
            style="sexo",
            errorbar="sd",
            ax=ejes[0])

sns.scatterplot(data=pinguinos,
            x="longitud_pico",
            y="longitud_aleta",
            hue="especie",
            s=300,
            ax=ejes[1])

plt.show()
```
