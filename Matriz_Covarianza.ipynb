La matriz de covarianza Σ de un vector aleatorio X de dimension n es una matriz n x n donde cada elemento Σij representa la covarianza entre las i-esimas y j-esimas componentes del vector:

Propiedades de la matriz de covarianza

- Simetrica: Σ=Σ^T
- Definidia no negativa: Para cualquier vector no nulo v, v^TΣv≥0
- Diagonalizable ortogonalmente: Dado que Σ es simetrica, puede ser diagonalizada por una matriz ortogonal Q, es decir, Σ = QAQ^T, donde A es una matriz diagonal de autovalores.

Calculo de la matriz de covarianza

La matriz de covarianza tambien puede expresarse como:

Σ = E[XX^T] - uu^T

donde u es el vector de medias de X y uu^T es el producto externo del vector de medias.

Transformaciones lineales de vectores aleatorios

Si Y = AX + B con A una matriz y B un vector constante, entonces la matriz de covarianza de Y se transforma segun:

ΣY = AΣxA^T

Esto es escencial para entender como las transformaciones lineales afectan las propiedades estadisiticas de los vectores aleatorios.

Decorrelar un vector aleatorio significa transformarlo de modo que su matriz de covarianza sea diagonal. Esto se logra a traves de la diagonalizacion ortogonal de Σ, utilizando los autovalores y autovectores de Σ. Las componentes del vector transformado seran independientes en el caso de distribuciones gaussianas.

Creacion y Propiedades de la matriz de Covarianza: Vamos a crear una matriz de covarianza y verificar sus propiedades basicas como la simetria y la definicion no negativa.

```python
import numpy as np

# Crear una matriz de covarianza de ejemplo
cov_matrix = np.array([[2, 1.7], [1.7, 1]])

# Verificar si la matriz es simétrica
print("Matriz de Covarianza:")
print(cov_matrix)
print("Es simétrica:", np.array_equal(cov_matrix, cov_matrix.T))

# Verificar la definición no negativa mediante los autovalores
eigenvalues = np.linalg.eigvals(cov_matrix)
print("Autovalores de la matriz de covarianza:", eigenvalues)
print("Todos los autovalores son no negativos:", np.all(eigenvalues >= 0))
```

Transformacion Lineal de variables aleatorias. Ilustraremos como una transformacion lineal afecta la matriz de covarianza.

```python
# Definir una matriz de transformación A y un vector B
A = np.array([[1, 0], [0, 2]])
B = np.array([1, -1])

# Nueva matriz de covarianza después de la transformación
transformed_cov_matrix = A @ cov_matrix @ A.T

print("Matriz de covarianza transformada:")
print(transformed_cov_matrix)
```

Decorrelacion de variables aleaatorias: Decorrelar un conjunto de variables aleatorias implica transformarlas para que su matriz de covarianza sea diagonal. Este proceso se realiza utilizando los autovalores de la matriz de covarianza.

```python
# Calcular los autovectores y autovalores
eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)

# Decorrelar usando la matriz de autovectores
decorrelation_matrix = eigenvectors.T @ cov_matrix @ eigenvectors

print("Matriz de covarianza decorrelada:")
print(decorrelation_matrix)
```

Verificacion de matrices de covarianza: Determinar si una matriz arbitraria puede ser una matriz de covarianza

```python
# Matriz arbitraria
M = np.array([[2, -4, 0], [-4, 3, 1], [0, 1, 2]])

# Verificar si M es simétrica y definida no negativa
print("M es simétrica:", np.array_equal(M, M.T))
print("Autovalores de M:", np.linalg.eigvals(M))
print("M es definida no negativa:", np.all(np.linalg.eigvals(M) >= 0))
```

```python
#Ejemplo Ej 3

import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la distribución
mu = np.array([0, 0])  # Media nula
cov_matrix = np.array([[2, 1.7], [1.7, 1]])  # Matriz de covarianza

# Generar muestras
samples = np.random.multivariate_normal(mu, cov_matrix, 1000)

# Estimar el vector media y la matriz de covarianza
estimated_mu = np.mean(samples, axis=0)
estimated_cov = np.cov(samples, rowvar=False)

print("Estimado mu:", estimated_mu)
print("Estimada matriz de covarianza:", estimated_cov)

# Decorrelación
eigvals, eigvecs = np.linalg.eigh(estimated_cov)
A = eigvecs @ np.diag(1.0 / np.sqrt(eigvals))
decorrelated_samples = (A @ (samples - estimated_mu).T).T

# Gráfico de las muestras decorrelacionadas
plt.figure(figsize=(8, 6))
plt.scatter(decorrelated_samples[:, 0], decorrelated_samples[:, 1], alpha=0.5)
plt.title("Muestras Decorrelacionadas")
plt.xlabel("$Y_1$")
plt.ylabel("$Y_2$")
plt.grid(True)
plt.show()
```

Forma cuadratica y contornos de densidad de una distribucion Gaussiana

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

def correct_covariance_matrix(matrix):
    """ Asegura que la matriz sea simétrica y positiva semidefinida. """
    # Hacer simétrica
    sym_matrix = (matrix + matrix.T) / 2
    # Corregir para que sea positiva semidefinida mediante autovalores
    eigvals, eigvecs = np.linalg.eigh(sym_matrix)
    corrected_eigvals = np.clip(eigvals, a_min=0, a_max=None)  # Elimina valores negativos
    return eigvecs @ np.diag(corrected_eigvals) @ eigvecs.T

# Parámetros de la distribución gaussiana bivariada
mu = np.array([0, 0])  # Vector de medias
cov_matrix = np.array([[2, 1.7], [1.7, 1]])  # Matriz de covarianza original

# Corregir la matriz de covarianza
corrected_cov_matrix = correct_covariance_matrix(cov_matrix)

# Generar una malla de puntos
x, y = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))
pos = np.dstack((x, y))

# Calcular la densidad de la distribución normal multivariable
rv = multivariate_normal(mu, corrected_cov_matrix, allow_singular=True)
z = rv.pdf(pos)

# Gráfico de contornos
plt.contourf(x, y, z, levels=30, cmap='viridis')
plt.colorbar()
plt.title('Contornos de Igual Densidad de una Gaussiana Bivariada')
plt.xlabel('X1')
plt.ylabel('X2')
plt.axis('equal')
plt.show()

```

Suma de Gaussianas y su varianza. Se desmotrara como la suma de gaussianas tambien es gaussiana y calcuaremos su varianza

```python
# Definir las variables aleatorias gaussianas
N = 10000
X1 = np.random.normal(0, 2, N)  # Media 0, desviación estándar 2
X2 = np.random.normal(0, 3, N)  # Media 0, desviación estándar 3

# Suma de las variables aleatorias
Y = X1 + X2

# Calcular media y varianza empíricas
media_Y = np.mean(Y)
varianza_Y = np.var(Y)

print(f"Media de Y: {media_Y}")
print(f"Varianza de Y: {varianza_Y}")

# Histograma de Y
plt.hist(Y, bins=50, alpha=0.7, color='blue', density=True)
plt.title('Histograma de la Suma de Dos Gaussianas')
plt.xlabel('Y')
plt.ylabel('Densidad')
plt.show()
```

Funcion caracteristica de una suma de Gaussianas: Demostrar como la funcion caracteristica puede ser calculada para una suma de variables aleatorias gaussianas

```python
# Calcular la función característica para una suma de gaussianas
def char_func_gaussian(t, mu, sigma_squared):
    return np.exp(1j * t * mu - 0.5 * sigma_squared * t**2)

t_values = np.linspace(-1, 1, 400)
phi_Y = char_func_gaussian(t_values, 0, 13)  # Suponiendo que la varianza combinada es 13

plt.plot(t_values, np.real(phi_Y), label='Parte Real')
plt.plot(t_values, np.imag(phi_Y), label='Parte Imaginaria')
plt.title('Función Característica de la Suma de Gaussianas')
plt.xlabel('t')
plt.ylabel('Función Característica')
plt.legend()
plt.grid(True)
plt.show()
```
