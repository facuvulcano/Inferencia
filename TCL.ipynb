TCL

Es uno de los conceptos mas importantes en estadistica. Este teorema establece que, bajo ciertas condiciones, la suma de una gran cantidad de variables aleatorias independientes y distribuidas identicamente (i.i.d) tendera a seguir una distribucion normal, independientemente de la forma de la distribucion original de las variables.

Detlles del TCL

Se aplica cuando se cumplen las siguientes condiciones:

- Las variables deben ser independientes
- Las variables deben estar distribuidas de manera identica
- La cantidad de variables, n, debe ser grande.

Formalmente, si X1, X2, ... , Xn son variables aleatorias i.i.d. con media mu y varianza sigma^2, entonces la suma Sn = X1 + X2 + ... + Xn normalizada por n (es decir, (sn - n*u) / (sigma * (raiz de n))) convergera en distribucion a una variable aleatoria normal estandar N(0, 1) a medida que n tiende a infinito.

Ejemplo en Python

Podemos ilustrar el TCL con un ejemplo en Python. Supongamos que queremos ver como la suma de variables aleatorias uniformes coverge a una distribucion normal. Podriamos hacerlo de la siguiente manera:

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el número de variables aleatorias en cada suma
n = 50

# Definir el número de sumas
num_sumas = 10000

# Generar sumas de variables uniformes
sumas = [np.sum(np.random.uniform(low=-0.5, high=0.5, size=n)) for _ in range(num_sumas)]

# Normalizar las sumas para demostrar el TCL
normalizadas = (sumas - np.mean(sumas)) / np.std(sumas)

# Graficar el histograma de las sumas normalizadas
plt.hist(normalizadas, bins=30, density=True, alpha=0.6, color='g')

# Superponer la densidad normal teórica
from scipy.stats import norm
x = np.linspace(min(normalizadas), max(normalizadas), 100)
plt.plot(x, norm.pdf(x), 'k', linewidth=2)
title = "Histograma de la suma normalizada de 50 variables uniformes"
plt.title(title)
plt.show()
```
