La asimetria, o "Skewness", es una medida estadistica que describe la simetria o la falta de simetria de una distribucion. Una distribucion simetrica tendra una asimetria de cero, minetras que una distribucion asimetrica tendra una asimetria que difiere de cero.

Caracteristicas de la Asimetria:

1- Asimetria Positiva (o sesgo a la derecha): En una distribucion con asimetria positiva, la cola en el lado derecho de la distribucion (los valores mas altos) es mas larga o mas extendida que la cola en el lado izquierdo, con una cola larga hacia la derecha.

2- Asimetria Negativa (o sesgo a la izquierda): En una disitribucion con asimetria negativa, la cola en el lado izquierdo de la distribucion (los valores mas bajos) es mas larga o mas extendida que la cola en el lado derecho. Esto indica que el promedio y los valores mas frecuentes estan mas hacia la derecha, con una cola larga hacia la izquierda.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, skew

# Generar datos de una distribución normal
data_normal = norm.rvs(size=1000, loc=0, scale=1)  # media=0, desviación estándar=1
media_normal = np.mean(data_normal)
varianza_normal = np.var(data_normal)
asimetria_norm = skew(data_normal)

print("Media de la distribución normal:", media_normal)
print("Varianza de la distribución normal:", varianza_normal)
print("Asimetría de la distribución exponencial:", asimetria_norm)

# Histograma de los datos
plt.hist(data_normal, bins=30, alpha=0.7, color='blue')
plt.title('Histograma de la Distribución Normal')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative

# Definir la función característica de una distribución normal estándar
def char_function_normal(omega):
    return np.exp(-omega**2 / 2)

# Rango de valores de omega para graficar
omega_values = np.linspace(-2, 2, 400)

# Calcular la función característica para esos valores
phi_values = char_function_normal(omega_values)

# Calcular la primera y segunda derivada evaluadas en omega = 0
first_derivative = derivative(char_function_normal, 0.0, dx=1e-6, n=1)
second_derivative = derivative(char_function_normal, 0.0, dx=1e-6, n=2)

# Media y varianza calculadas a partir de las derivadas
mean_calculated = -1j * first_derivative
variance_calculated = -second_derivative + (-1j * first_derivative)**2

# Mostrar resultados
print(f"Media calculada: {mean_calculated}")
print(f"Varianza calculada: {variance_calculated}")

# Graficar la función característica
plt.plot(omega_values, phi_values.real, label='Parte Real')
plt.plot(omega_values, phi_values.imag, label='Parte Imaginaria')
plt.title('Función Característica de la Distribución Normal Estándar')
plt.xlabel('Omega (ω)')
plt.ylabel('Función Característica Φ(ω)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.misc import derivative

def char_function_X(omega):
    """Calcula la función característica para la variable aleatoria X."""
    # Definir los valores y probabilidades de X
    values = np.array([1, 2, 3])
    probabilities = np.array([0.2, 0.5, 0.3])
    
    # Calcular la función característica
    phi_X = np.sum(probabilities * np.exp(1j * omega * values))
    return phi_X


# Calcular la primera y segunda derivada en omega = 0
first_derivative = derivative(char_function_X, 0.0, dx=1e-6, n=1)
second_derivative = derivative(char_function_X, 0.0, dx=1e-6, n=2)

# Calcular la media y la varianza
media = first_derivative.imag  # Parte imaginaria de la primera derivada
varianza = second_derivative.real - media**2  # Real de la segunda - media^2

print("Media calculada:", media)
print("Varianza calculada:", varianza)
```

```python
import numpy as np
from scipy.misc import derivative

def char_function_example(omega):
    # Esta función debe ser cambiada a la función característica real de tu variable aleatoria.
    return np.exp(-omega**2 / 2)  # Ejemplo para una distribución normal estándar.

# Calcular las primeras tres derivadas en omega = 0
moments = [derivative(char_function_example, 0.0, dx=1e-6, n=i, order=2*i+1) for i in range(1, 4)]
print("Primeros tres momentos:", moments)
```

```python
def char_function_geometric(omega, p):
    # Utilizando la fórmula de la función característica para la distribución geométrica
    return p / (1 - (1 - p) * np.exp(1j * omega))

# Calcula la media y la varianza utilizando las derivadas de la función característica
media = derivative(lambda x: char_function_geometric(x, 0.5).imag, 0.0, dx=1e-6, n=1)
varianza = derivative(lambda x: char_function_geometric(x, 0.5).real, 0.0, dx=1e-6, n=2) - media**2

print("Media:", media)
print("Varianza:", varianza)
```
